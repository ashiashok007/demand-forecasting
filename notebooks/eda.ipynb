```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/synthetic_data.csv', parse_dates=['date','expiry_date'])
df.head()
```

```python
df.describe(include='all').T   # Summary statistics for numeric & categorical
```

```python
df.isna().sum()            # Count of nulls per column
```

Visualize Key Patterns
a. Sales Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['units_sold'], bins=30, kde=True)
plt.title('Units Sold Distribution')
plt.show()
```

b. Monthly Trend

```python
monthly_sales = df.set_index('date').resample('M')['units_sold'].sum()
monthly_sales.plot(marker='o')
plt.title('Monthly Units Sold')
plt.ylabel('Total Units')
plt.show()
```

c. Stock-Out Frequency by Store

```python
stockout_rates = df.groupby('store_id')['stock_out_flag'].mean().sort_values(ascending=False)
print(stockout_rates)
```

```python
stockout_rates = df.groupby('store_id')['stock_out_flag'].mean().sort_values(ascending=False)
sns.barplot(x=stockout_rates.index, y=stockout_rates.values)
plt.title('Stock-Out Rate by Store')
plt.ylabel('Rate')
plt.show()
```

d. Days to Expiry by Category

```python
df['days_to_expiry'] = (df['expiry_date'] - df['date']).dt.days
sns.boxplot(x='product_category', y='days_to_expiry', data=df)
plt.title('Days to Expiry by Category')
plt.show()
```
